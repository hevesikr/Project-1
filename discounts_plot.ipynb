```python
orderlines_products = orderlines_qu.merge(products_cl, how="left", on = "sku")[["category","price","unit_price" ]]
orderlines_products#.sample(10)
```

```python
orderlines_products["discount"] = orderlines_products["price"] - orderlines_products["unit_price"]

orderlines_products
```

```python
orderlines_products.sort_values("discount", ascending=False)
```

```python
first_fifteen = orderlines_products.groupby('category').agg(discount=('discount', 'mean')).reset_index()
first_fifteen = first_fifteen.sort_values(by='discount', ascending=False).head(15)
first_fifteen
```

```python
order = first_fifteen.groupby("category")["discount"].mean().sort_values(ascending=False).index
order
```

```python
my_colors_cat2 = {
   category: ("turquoise" if category in ["desktop", "expansion kit", "monitor", "support/stand/docking station", "graphics tablet"] else "beige")
    for category in first_fifteen['category'].unique()
}

my_colors_cat2
```

```python
discounts_plot = sns.catplot(kind="bar",
            data=first_fifteen,
            x='category',
            y = 'discount',
            hue ='category',
            palette = my_colors_cat2,
            height=6,
            aspect=6,
            width=1,
            order=order
            );


discounts_plot.set_xticklabels(fontsize=26, rotation=45, ha='right')
discounts_plot.set_yticklabels(fontsize=26)
discounts_plot.ax.set_title("Discounts by categories",
                     fontsize=36)
discounts_plot.set_axis_labels("Categories", "Discounts in €", fontsize=28);
```
